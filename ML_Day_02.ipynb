```python
# Types of Machine Learning

'''
Based on how much supervision needed, we can classify ML in 4 types:
1. Supervised Learning (Regression, Classificaiton)
2. Unsupervised Learning (Clustering, Dimensionality Reduction, Anamoly Detection, Association)
3. Reinforcement Learning
4. Semi-supervised Learning


# Supervised Learning
  Supervised learning is a type of machine learning where an AI model is trained using labeled data sets, meaning every input example is paired with its correct corresponding output answer.
  We just need to find the relation between input and output so that we can predict new output for a new input.
  Examples: Regression, Classification
  Regression - Predicts something (Output column is numberical)
  Classification - Predict class, Categorize something (Output column is categorical)



# Unsupervised Learning
  Unsupervised learning is a branch of machine learning where an algorithm analyzes unlabeled datasets to discover hidden patterns, structures, and relationships without human intervention or guidance.
  Unlike supervised learning, which relies on predefined pairs of inputs and outputs, unsupervised learning works on raw data to find similarities or differences completely on its own
  Examples: Clustering, Dimensionality Reduction, Anamoly Detection, Association Rule Learning
  Clustering - Can categorize data
  Dimensionality Reduction - Reduce unnecessary columns/dimensions by combining multiple columns in a single column
  Anamoly Detection - detect outlier and remove them from the system
  Association Rule Learning - Association rules are a fundamental concept used to find relationships, correlations or patterns within large sets of data items.a rule-based unsupervised machine learning technique used to discover hidden, "If-Then" relationships between variables in large databases.
                              Commonly used for Market Basket Analysis, it reveals patterns like "customers who buy bread also buy butter" to optimize product placement and recommendations.


# Semi-supervised Learning
  Semi-supervised learning is a distinct machine learning approach that uses a small amount of labeled data along with a large amount of unlabeled data to improve model performance. The goal is to learn a function that accurately predicts outputs based on inputs, similar to supervised learning, but with much less labelled data.
  It bridges the gap between supervised and unsupervised learning, making it ideal when manual data annotation is expensive or time-consuming.(https://www.geeksforgeeks.org/machine-learning/ml-semi-supervised-learning/)

# Reinforcement Learning
  Reinforcement learning (RL) is a machine learning method where an AI "agent" learns to make decisions through trial and error. By interacting with a dynamic environment, the agent takes actions and receives feedback in the form of rewards or penalties, adjusting its behavior to maximize its cumulative reward over time.
  Core components: Environment, Agent, Reward, State, Action


** When we develop a ML model and it runs on a server then the server is called production environment in machine leraning**
** When we code it on our local machine then it is called development environment**

We can classify ML based on production environment:
1. Batch Learning / Offline Learning
2. Online Learning

1. Batch Learning / Offline Learning
  Batch machine learning (or offline learning) is an approach where a model is trained using an entire dataset at once, rather than incrementally. The model crunches the collected data offline, gets deployed, and then remains static until it is retrained on a newer dataset.
  (https://www.geeksforgeeks.org/artificial-intelligence/batch-offline-learning-vs-online-learning-in-artificial-intelligence/)

  What we do in batch learning is, we take entire data and train the model in our local computer offline and then deploy it in a server. As a result, it being a static model where new updates not available instantly.
  For update new data, we need to pull back our model, re-train our model based on new data and then deploy it. So, batch learning fails here if we want updation instantly.

2.




'''
```
