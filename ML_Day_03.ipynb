```python
'''
# Online Machine Learning
    Online machine learning is an AI method where a model incrementally learns from a continuous stream of real-time data, updating its predictions step-by-step rather than retraining on a massive dataset all at once.
  It is ideal for environments that require instant adaptation to new trends. We can train our model when it is already in server(online). That's why this is called online learning.

  *When to use*:
    - Where there is a concept drift(problem statement or cocept changes/evolvs with time)
    - Cost effection
    - Faster solution


# Based on how our machine learns we can divide it into two types:
 1. Instance Based Learning (Memorizing)
 2. Model Based Learning (Generalization/Understands)

 1. Instace Based Learning
      Instance-based learning is a machine learning approach where the system memorizes the training dataset instead of building an explicit mathematical model.
      When a new, unseen data point is introduced, the system makes a prediction by comparing it directly to the stored training examples using a similarity metric.
      In this learning our system directly memorizes everything without understanding and make prediction based on comparison.

 2. Model Based Learning
      Model-based learning is a machine learning approach where a system extracts generalized patterns from training data to construct an explicit mathematical model or set of rules. Once this model is built, the original training data can be entirely discarded; the system makes all future predictions using only the learned parameters of the model.
      This stands in direct contrast to instance-based learning (like K-Nearest Neighbors), which memorizes raw data points and calculates similarities at runtime.
      In this learning our system doesn't directly memorizes the data  only, it tries to extract a pattern from data and predict based on that pattern after learning from data.


  *core difference between instance based and model based learning*
    -Instance-based learning memorizes training data and makes predictions by comparing new inputs to these stored examples.
    In contrast, model-based learning extracts general patterns from data to build a mathematical model, which is then used to make predictions without referencing the original data.
    In model based learning, it requires less storage to store a model whether instance based learning require more storage to store dataset.



# Challenges in ML:
  1. Data collection (Fetch data from API or web scraping)
  2. Insufficient data or Labelled data
  3. Non-representative data
  4. Poor quality data
  5. Irrelevant features
  6. Overfitting (Memorizes data)
  7. Underfitting (not memorize thoroughly and also not generalize thoroughly)
  8. Software integration
  9. Offline learning/Deployment
  10. Cost involved



**
Web scraping:
  Web scraping is the automated process of using software or bots to extract unstructured public data from websites.
  It eliminates the need for manual copy-pasting, gathering raw HTML or APIs and transforming the content into clean, structured formats like spreadsheets or databases for analysis.

Overfitting and Underfitting:
  Instance-based learning memorizes training data and makes predictions by comparing new inputs to these stored examples.
  In contrast, model-based learning extracts general patterns from data to build a mathematical model, which is then used to make predictions without referencing the original data
**
'''
```
